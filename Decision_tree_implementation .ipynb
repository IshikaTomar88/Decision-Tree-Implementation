# Decision Tree Classifier on Iris Dataset

Obective: demonstrates how to build, train, and visualize a Decision Tree Classifier using the Iris dataset.

#### Steps:

*   Import necessary libraries for data manipulation, visualization, and machine learning.
*   Load and explore the Iris dataset.
*   Split the dataset into training and testing sets.
*   Build and train a Decision Tree Classifier model.
*   Make predictions using the trained model.
*   Visualize the decision tree to understand the model's decision rules.
*   Save the decision tree visualization and a text representation of the tree.


# Import Libraries

This section imports all the necessary libraries for data manipulation, visualization, and machine learning.

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import tree
```

## Load the Dataset

Here we load the Iris dataset, which is a classic dataset used in machine learning for classification tasks. It contains measurements of iris flowers and their corresponding species.

```python
# Loading iris dataset
from sklearn.datasets import load_iris
iris = load_iris()

# Create a pandas DataFrame for features
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Create a pandas Series for target variable
y = pd.Series(iris.target)
```

## Explore the Dataset

Before proceeding with the model training, it's good practice to explore the dataset to understand its structure and content.

```python
print(X.head())
print(y.value_counts())
```

## Split the Dataset

We split the dataset into training and testing sets to evaluate the model's performance on unseen data. The training set is used to train the model, and the testing set is used to assess its accuracy.

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

## Check the Shape of the Datasets

Let's check the dimensions of the training and testing sets to ensure the split was performed correctly.

```python
# check the shape of the X_train, X_test, y_train, y_test
print("X train shape:", X_train.shape)
print("X test shape:", X_test.shape)
print("y train shape:", y_train.shape)
print("y test shape:", y_test.shape)
```

## Build and Train the Decision Tree Model

Here we create an instance of the Decision Tree Classifier and train it on the training data. The model learns the decision rules from the training data to classify the iris species.

```python
# Make an instance of the Decision Tree Classifier model
from sklearn.tree import DecisionTreeClassifier
clf_model = DecisionTreeClassifier()
```

```python
# Train the model on the training data
# X_train: training features
# y_train: training target variable
clf_model.fit(X_train, y_train)
```

## Make Predictions

Use the trained model to make predictions on the test set.

```python
# Predict the labels of the unseen (test) data
# X_test: testing features
y_pred = clf_model.predict(X_test)
```

## Visualize the Decision Tree

Visualize the trained Decision Tree model. This helps in understanding the decision rules learned by the model.

```python
# Create a figure for plotting the decision tree
fig = plt.figure()

# Plot the decision tree
tree.plot_tree(clf_model)

# Set the title of the plot
plt.title('Decision Tree Visualization')

# Display the plot
plt.show()
```

## Save the Decision Tree Visualization

Save the decision tree plot to a file for later use or sharing.

```python
# Save the decision tree plot to a file
# 'decision_tree.png': the filename to save the plot
fig.savefig('decision_tree.png')

# Display the plot
display(fig)
```

# Visualize A Decision Tree (Alternative Method)

This section demonstrates an alternative method to visualize a Decision Tree using the entire dataset.

## Load the dataset (Alternative Method)

Load the dataset again for the alternative visualization method.

```python
# Load the dataset
iris = load_iris()
```

## Assign features and target (Alternative Method)

Assign features and target variables for the alternative visualization method.

```python
# Assign features to X and target to y
X = iris.data
y = iris.target
```

## Train Decision Tree (Alternative Method)

Train a Decision Tree model on the entire dataset for the alternative visualization method.

```python
# Create a decision tree classifier object
clf_model = DecisionTreeClassifier()
```

```python
# Train the model on the entire dataset
# X: features
# y: target variable
clf_model.fit(X, y)
```

## Export Text Representation of the Decision Tree

Export the decision tree as a text representation. This provides a textual representation of the decision rules.

```python
# Export the decision tree as a text representation
text_representation = tree.export_text(clf_model)

# Print the text representation of the decision tree
print(text_representation)
```

## Save the Text Representation

Save the text representation of the decision tree to a file.

```python
# Save the text representation of the decision tree to a file
# "decision_tree.log": the filename to save the text representation
# "w": opens the file in write mode
with open("decision_tree.log", "w") as fout:
    fout.write(text_representation)
```

## Visualize the Decision Tree with Enhanced Options

Visualize the decision tree with enhanced options like filled nodes, feature names, and class names. This provides a more informative visualization of the tree.

```python
fig = plt.figure(figsize=(15, 10))
tree.plot_tree(clf_model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)

# Display the plot
display(fig)
```
